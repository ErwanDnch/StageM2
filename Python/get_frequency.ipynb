```python
from cycler import cycler
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.gridspec import GridSpec
import pandas as pd

import matplotlib
```

```python
df = pd.read_csv("data_o0.0_phi0_om0.csv")
phi = list(df["phi"])
dt = 0.05

plt.plot(phi)
```

```python
phi = phi[int(30/dt):]
plt.plot(phi)
```

```python
Y = np.fft.fft(phi)
freqs = np.fft.fftfreq(len(phi), dt)
```

```python
# Take only positive frequencies (real signals are symmetric in FFT)
positive_freqs = freqs[:len(phi)//2]
positive_magnitude = np.abs(Y[:len(phi)//2])

# Find the frequency with max magnitude (dominant frequency)
peak_idx = np.argmax(positive_magnitude)
dominant_freq = positive_freqs[peak_idx]

print(f"Dominant frequency = {dominant_freq:.4f} Hz")

df = positive_freqs[1] - positive_freqs[0]
```

```python
plt.bar(positive_freqs, positive_magnitude, width=df)
plt.xlabel('Frequency (Hz)')
plt.title('Frequency spectrum')
plt.grid()
plt.xlim(-0,1)
plt.show()
```
